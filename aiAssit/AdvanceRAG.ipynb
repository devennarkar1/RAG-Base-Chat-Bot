```python
# ─────────────────────────────────────────────────────────
# 2. METADATA ENRICHMENT
#    Adds year, category, author for Self-Query filtering
# ─────────────────────────────────────────────────────────
FILE_METADATA = {
    "Handbook Of Natural Language Processing": {
        "year": 2010, "category": "NLP", "author": "Indurkhya & Damerau"
    },
    "NAACL2013-Socher-Manning-DeepLearning": {
        "year": 2013, "category": "Deep Learning", "author": "Socher & Manning"
    },
    "NLTK": {
        "year": 2009, "category": "NLP Toolkit", "author": "Bird, Klein & Loper"
    },
    "Natural Language Processing Recipes": {
        "year": 2019, "category": "Applied NLP", "author": "Ekbal"
    },
    "Natural Language Processing-1": {
        "year": 2020, "category": "NLP", "author": "Unknown"
    },
}

def enrich_metadata(doc: Document, source_file: str) -> Document:
    """Adds year, category, author metadata to each document chunk."""
    doc.metadata["source_file"] = source_file
    for key, meta in FILE_METADATA.items():
        if key in source_file:
            doc.metadata.update(meta)
            break
    else:
        doc.metadata.setdefault("year", 0)
        doc.metadata.setdefault("category", "Unknown")
        doc.metadata.setdefault("author", "Unknown")
    return doc

```

```python
# ─────────────────────────────────────────────────────────
# 3. DOCUMENT LOADING
# ─────────────────────────────────────────────────────────
def load_documents(pdf_files: List[str]) -> List[Document]:
    documents = []
    for file in pdf_files:
        print(f"  📄 Loading: {os.path.basename(file)}")
        loader = PyPDFLoader(file)
        docs = loader.load()
        source_file = os.path.basename(file)
        docs = [enrich_metadata(doc, source_file) for doc in docs]
        documents.extend(docs)
    print(f"  ✅ Loaded {len(documents)} pages total.\n")
    return documents
```

```python
def build_splitters(embeddings):
    """
    Returns (semantic_splitter, pdr_child_splitter, parent_splitter).

    semantic_splitter    : SemanticChunker for standalone FAISS index
                           (best quality, topic-aware boundaries)
    pdr_child_splitter   : RecursiveCharacterTextSplitter for
                           ParentDocumentRetriever (Pydantic-compatible)
    parent_splitter      : Large chunks returned to LLM for full context
    """
    # Always create the Pydantic-safe recursive splitter
    pdr_child_splitter = RecursiveCharacterTextSplitter(
        chunk_size=CHILD_CHUNK_SIZE,
        chunk_overlap=CHILD_CHUNK_OVERLAP,
        add_start_index=True,
    )

    if SEMANTIC_CHUNKER_AVAILABLE:
        print("  ✅ SemanticChunker ready for FAISS index (topic-aware boundaries).")
        print("  ℹ️  ParentDocumentRetriever uses RecursiveCharacterTextSplitter (Pydantic requirement).")
        semantic_splitter = SemanticChunker(
            embeddings,
            breakpoint_threshold_type="percentile",
            breakpoint_threshold_amount=85,
        )
    else:
        print("  ⚠️  SemanticChunker unavailable — using RecursiveCharacterTextSplitter for both.")
        semantic_splitter = pdr_child_splitter  # Same object, no penalty

    parent_splitter = RecursiveCharacterTextSplitter(
        chunk_size=PARENT_CHUNK_SIZE,
        chunk_overlap=PARENT_CHUNK_OVERLAP,
        add_start_index=True,
    )
    return semantic_splitter, pdr_child_splitter, parent_splitter


  # # ─────────────────────────────────────────────────────────
# # 4. UPGRADE 4 — SEMANTIC CHUNKING
# #    Breaks chunks at topic boundaries, not character counts.
# #    Falls back to RecursiveCharacterTextSplitter if unavailable.
# # ─────────────────────────────────────────────────────────
# def build_splitters(embeddings):
#     """Returns (child_splitter, parent_splitter)."""
#     if SEMANTIC_CHUNKER_AVAILABLE:
#         print("  ✅ Using SemanticChunker for child chunks.")
#         child_splitter = SemanticChunker(
#             embeddings,
#             breakpoint_threshold_type="percentile",  # Breaks on topic change
#             breakpoint_threshold_amount=85,
#         )
#     else:
#         print("  ⚠️  Using RecursiveCharacterTextSplitter for child chunks.")
#         child_splitter = RecursiveCharacterTextSplitter(
#             chunk_size=CHILD_CHUNK_SIZE,
#             chunk_overlap=CHILD_CHUNK_OVERLAP,
#             add_start_index=True,
#         )

#     parent_splitter = RecursiveCharacterTextSplitter(
#         chunk_size=PARENT_CHUNK_SIZE,
#         chunk_overlap=PARENT_CHUNK_OVERLAP,
#         add_start_index=True,
#     )
#     return child_splitter, parent_splitter
```

```python
# ─────────────────────────────────────────────────────────
# 5. UPGRADE 1 — HYBRID SEARCH (BM25 + FAISS)
#    BM25 catches exact keyword matches (e.g. "TF-IDF", "BERT").
#    FAISS catches semantic meaning.
#    EnsembleRetriever merges both with configurable weights.
# ─────────────────────────────────────────────────────────
def build_hybrid_retriever(
    all_chunks: List[Document],
    vector_store: FAISS,
    k: int = TOP_K_RETRIEVAL,
    bm25_weight: float = 0.4,
    faiss_weight: float = 0.6,
) -> EnsembleRetriever:
    """
    bm25_weight: Higher = more keyword matching
    faiss_weight: Higher = more semantic matching
    They must sum to 1.0
    """
    bm25_retriever = BM25Retriever.from_documents(all_chunks)
    bm25_retriever.k = k

    faiss_retriever = vector_store.as_retriever(search_kwargs={"k": k})

    hybrid_retriever = EnsembleRetriever(
        retrievers=[bm25_retriever, faiss_retriever],
        weights=[bm25_weight, faiss_weight],
    )
    print(f"  ✅ Hybrid Retriever: BM25({bm25_weight}) + FAISS({faiss_weight})")
    return hybrid_retriever
```

```python
# ─────────────────────────────────────────────────────────
# 6. UPGRADE 3 — CROSS-ENCODER RE-RANKING
#    After retrieval, a Cross-Encoder scores every (query, doc) pair
#    and re-orders by true relevance before sending to LLM.
#    Model: BAAI/bge-reranker-base (free, runs locally)
# ─────────────────────────────────────────────────────────
class CrossEncoderReranker:
    def __init__(self, model_name: str = "BAAI/bge-reranker-base", top_n: int = TOP_K_FINAL):
        if RERANKER_AVAILABLE:
            self.model = CrossEncoder(model_name)
            self.top_n = top_n
            print(f"  ✅ Cross-Encoder Re-ranker loaded: {model_name}")
        else:
            self.model = None
            self.top_n = top_n

    def rerank(self, query: str, documents: List[Document]) -> List[Document]:
        if self.model is None or not documents:
            return documents[:self.top_n]

        pairs = [(query, doc.page_content) for doc in documents]
        scores = self.model.predict(pairs)

        scored_docs = sorted(
            zip(scores, documents),
            key=lambda x: x[0],
            reverse=True
        )
        reranked = [doc for _, doc in scored_docs[:self.top_n]]
        print(f"  🔀 Re-ranked {len(documents)} → kept top {self.top_n} docs")
        return reranked
```

```python
# ─────────────────────────────────────────────────────────
# 7. UPGRADE 6 — HyDE (Hypothetical Document Embeddings)
#    Instead of embedding the raw query, the LLM first generates
#    a "fake ideal answer", which is then used for semantic search.
#    This dramatically improves recall for short/vague queries.
# ─────────────────────────────────────────────────────────
HYDE_PROMPT = ChatPromptTemplate.from_template("""
You are an expert in Natural Language Processing and Machine Learning.
Given the following question, write a detailed, factual paragraph (3-5 sentences)
that would appear in an academic textbook or research paper as the ideal answer.
Do not say "I think" or "The answer is". Just write the paragraph directly.

Question: {question}

Ideal answer paragraph:
""")

def build_hyde_query(query: str, llm) -> str:
    """Generates a hypothetical ideal answer to improve search recall."""
    chain = HYDE_PROMPT | llm | StrOutputParser()
    hypothetical_doc = chain.invoke({"question": query})
    print(f"\n  💡 HyDE generated hypothetical doc ({len(hypothetical_doc)} chars)")
    return hypothetical_doc
```

```python
# ─────────────────────────────────────────────────────────
# 8. UPGRADE 5 — SELF-QUERYING RETRIEVER
#    The LLM writes its own metadata filters automatically.
#    E.g., "Find BERT info only from papers after 2012" →
#         filter: year > 2012
# ─────────────────────────────────────────────────────────
METADATA_FIELD_INFO = [
    AttributeInfo(name="year", description="Publication year of the document", type="integer"),
    AttributeInfo(name="category", description="Category: NLP, Deep Learning, Applied NLP, NLP Toolkit", type="string"),
    AttributeInfo(name="author", description="Author(s) of the document", type="string"),
    AttributeInfo(name="source_file", description="The PDF filename", type="string"),
]

def build_self_query_retriever(vector_store: FAISS, llm) -> SelfQueryRetriever:
    return SelfQueryRetriever.from_llm(
        llm=llm,
        vectorstore=vector_store,
        document_contents="NLP and Machine Learning academic texts and research papers",
        metadata_field_info=METADATA_FIELD_INFO,
        verbose=False,
    )
```

```python
# ─────────────────────────────────────────────────────────
# 9. UPGRADE 7 — RAGAS EVALUATION FRAMEWORK
#    Scores your RAG pipeline on:
#    - Faithfulness: Did LLM hallucinate?
#    - Answer Relevancy: Is the answer on-topic?
#    - Context Precision: Is retrieved context actually useful?
#    - Context Recall: Did we retrieve everything needed?
# ─────────────────────────────────────────────────────────
def evaluate_with_ragas(
    questions: List[str],
    answers: List[str],
    contexts: List[List[str]],
    ground_truths: Optional[List[str]] = None,
):
    if not RAGAS_AVAILABLE:
        print("⚠️  Ragas not available. Skipping evaluation.")
        return None

    if ground_truths is None:
        ground_truths = [""] * len(questions)

    eval_dataset = Dataset.from_dict({
        "question": questions,
        "answer": answers,
        "contexts": contexts,
        "ground_truth": ground_truths,
    })

    metrics = [faithfulness, answer_relevancy, context_precision]
    if ground_truths and ground_truths[0]:
        metrics.append(context_recall)

    print("\n  📊 Running Ragas evaluation (this may take a minute)...")
    result = evaluate(eval_dataset, metrics=metrics)
    return result
```

```python
# ─────────────────────────────────────────────────────────
# 10. EXTRA UPGRADE — QUERY EXPANSION
#     Generates 3 alternative phrasings of the query.
#     All versions search in parallel → better recall.
# ─────────────────────────────────────────────────────────
QUERY_EXPANSION_PROMPT = ChatPromptTemplate.from_template("""
You are an AI assistant helping to improve document search.
Generate exactly 3 alternative phrasings of the following query.
Each rephrasing should approach the question from a different angle.
Return ONLY the 3 questions, one per line, no numbering or extra text.

Original query: {query}

3 alternative phrasings:
""")

def expand_query(query: str, llm) -> List[str]:
    """Returns original + 3 expanded queries."""
    chain = QUERY_EXPANSION_PROMPT | llm | StrOutputParser()
    expanded = chain.invoke({"query": query})
    variants = [q.strip() for q in expanded.strip().split("\n") if q.strip()][:3]
    all_queries = [query] + variants
    print(f"  🔄 Query expanded to {len(all_queries)} variants")
    return all_queries
```

```python
# ─────────────────────────────────────────────────────────
# 11. EXTRA UPGRADE — CONTEXTUAL COMPRESSION
#     Removes irrelevant sentences from retrieved chunks
#     before sending to LLM, reducing noise & token waste.
# ─────────────────────────────────────────────────────────
def build_compressed_retriever(base_retriever, llm):
    try:
        from langchain.retrievers import ContextualCompressionRetriever
        from langchain.retrievers.document_compressors import LLMChainExtractor
        compressor = LLMChainExtractor.from_llm(llm)
        compressed = ContextualCompressionRetriever(
            base_compressor=compressor,
            base_retriever=base_retriever,
        )
        print("  ✅ Contextual Compression enabled")
        return compressed
    except Exception as e:
        print(f"  ⚠️  Contextual Compression not available: {e}")
        return base_retriever
```

```python
# ─────────────────────────────────────────────────────────
# 12. MAIN ADVANCED RAG CLASS
# ─────────────────────────────────────────────────────────
class AdvancedRAGPipeline:
    """
    All-in-one Advanced RAG Pipeline with:
    ✅ Semantic Chunking
    ✅ Hybrid Search (BM25 + FAISS)
    ✅ Parent-Document Retrieval
    ✅ Cross-Encoder Re-ranking
    ✅ HyDE Query Enhancement
    ✅ Query Expansion
    ✅ Self-Query with Metadata Filtering
    ✅ Contextual Compression
    ✅ Ragas Evaluation
    ✅ Rich Source Citations
    """

    def __init__(self, pdf_files: List[str], groq_api_key: str):
        print("🚀 Initializing Advanced RAG Pipeline...\n")

        # ── LLM ──────────────────────────────────────────
        self.llm = ChatGroq(
            api_key=groq_api_key,
            model_name="llama-3.3-70b-versatile",
            temperature=0.1,
        )

        # ── Embeddings ───────────────────────────────────
        print("📦 Loading embeddings model...")
        self.embeddings = HuggingFaceEmbeddings(
            model_name="sentence-transformers/all-MiniLM-L6-v2"
        )

        # ── Load Documents ───────────────────────────────
        print("📚 Loading PDFs...")
        self.documents = load_documents(pdf_files)

        # ── Build Splitters (Semantic + Parent) ──────────
        print("✂️  Building splitters...")
        # self.child_splitter, self.parent_splitter = build_splitters(self.embeddings)
        self.semantic_splitter, self.pdr_child_splitter, self.parent_splitter = build_splitters(self.embeddings)


        # ── Split into child chunks for indexing ─────────
        print("🔪 Splitting documents into chunks...")
        self.child_chunks = self.semantic_splitter.split_documents(self.documents)
        print(f"  📊 Created {len(self.child_chunks)} child chunks")

        # ── Build FAISS Vector Store ──────────────────────
        print("🗄️  Building FAISS vector store...")
        self.vector_store = FAISS.from_documents(self.child_chunks, self.embeddings)

        # ── Upgrade 1: Hybrid Retriever ───────────────────
        print("\n⚙️  Setting up retrievers...")
        self.hybrid_retriever = build_hybrid_retriever(
            self.child_chunks, self.vector_store
        )

        # ── Upgrade 2: Parent-Document Retriever ──────────
        self.doc_store = InMemoryStore()
        self.parent_retriever = ParentDocumentRetriever(
            vectorstore=self.vector_store,
            docstore=self.doc_store,
            child_splitter=self.pdr_child_splitter, # Use pdr_child_splitter here
            parent_splitter=self.parent_splitter,
        )
        self.parent_retriever.add_documents(self.documents)
        print("  ✅ Parent-Document Retriever ready")

        # ── Upgrade 3: Cross-Encoder Re-ranker ────────────
        self.reranker = CrossEncoderReranker()

        # ── Upgrade 5: Self-Query Retriever ───────────────
        try:
            self.self_query_retriever = build_self_query_retriever(
                self.vector_store, self.llm
            )
            print("  ✅ Self-Query Retriever ready")
        except Exception as e:
            print(f"  ⚠️  Self-Query Retriever setup failed: {e}")
            self.self_query_retriever = None

        print("\n✅ Pipeline fully initialized!\n")

    # ──────────────────────────────────────────────────────
    # Core Query Method
    # ──────────────────────────────────────────────────────
    def query(
        self,
        question: str,
        use_hyde: bool = True,
        use_query_expansion: bool = True,
        use_parent_docs: bool = True,
        use_reranking: bool = True,
        use_self_query: bool = False,  # Set True for metadata-filtered queries
        verbose: bool = True,
    ) -> Dict[str, Any]:
        """
        Full advanced RAG query.

        Parameters:
            question: Your question
            use_hyde: Use HyDE to generate hypothetical doc for search
            use_query_expansion: Generate alternative query phrasings
            use_parent_docs: Retrieve parent (large) chunks for context
            use_reranking: Apply Cross-Encoder re-ranking
            use_self_query: Use metadata-filtered retrieval
        """
        if verbose:
            print(f"\n{'='*60}")
            print(f"❓ QUERY: {question}")
            print('='*60)

        retrieved_docs = []
        search_query = question

        # ── UPGRADE 6: HyDE ───────────────────────────────
        if use_hyde:
            if verbose: print("\n[1/5] 🧠 Generating HyDE hypothetical document...")
            search_query = build_hyde_query(question, self.llm)

        # ── UPGRADE Query Expansion ───────────────────────
        if use_query_expansion:
            if verbose: print("\n[2/5] 🔄 Expanding query variants...")
            query_variants = expand_query(question, self.llm)
        else:
            query_variants = [question]

        # ── UPGRADE 1+2: Hybrid + Parent Retrieval ────────
        if verbose: print("\n[3/5] 🔍 Retrieving documents...")

        if use_self_query and self.self_query_retriever:
            # Use metadata-filtered retrieval
            retrieved_docs = self.self_query_retriever.get_relevant_documents(question)
            if verbose: print(f"  Self-Query retrieved: {len(retrieved_docs)} docs")
        else:
            # Retrieve using all query variants, then deduplicate
            seen_contents = set()
            for q in query_variants:
                if use_parent_docs:
                    docs = self.parent_retriever.get_relevant_documents(q)
                else:
                    docs = self.hybrid_retriever.get_relevant_documents(q)

                for doc in docs:
                    content_hash = hash(doc.page_content[:200])
                    if content_hash not in seen_contents:
                        seen_contents.add(content_hash)
                        retrieved_docs.append(doc)

            if verbose: print(f"  Retrieved {len(retrieved_docs)} unique docs across {len(query_variants)} query variants")

        # ── UPGRADE 3: Re-Ranking ─────────────────────────
        if use_reranking and retrieved_docs:
            if verbose: print(f"\n[4/5] 🎯 Re-ranking {len(retrieved_docs)} documents...")
            final_docs = self.reranker.rerank(question, retrieved_docs)
        else:
            final_docs = retrieved_docs[:TOP_K_FINAL]

        # ── LLM Answer Generation ─────────────────────────
        if verbose: print("\n[5/5] 💬 Generating answer with LLM...")

        context = "\n\n---\n\n".join([doc.page_content for doc in final_docs])

        answer_prompt = ChatPromptTemplate.from_template("""
You are an expert assistant specializing in Natural Language Processing and Machine Learning.
Answer the question based ONLY on the provided context.
If the context doesn't contain enough information, say so clearly.
Be precise, cite specific concepts, and explain clearly.

Context:
{context}

Question: {question}

Detailed Answer:""")

        chain = answer_prompt | self.llm | StrOutputParser()
        answer = chain.invoke({"context": context, "question": question})

        # ── Build Result ──────────────────────────────────
        result = {
            "query": question,
            "answer": answer,
            "source_documents": final_docs,
            "num_docs_retrieved": len(retrieved_docs),
            "num_docs_used": len(final_docs),
        }

        if verbose:
            self._print_result(result)

        return result

    def _print_result(self, result: Dict[str, Any]):
        """Pretty-prints the query result with sources."""
        print(f"\n{'='*60}")
        print("✅ ANSWER:")
        print('='*60)
        print(result["answer"])
        print(f"\n{'─'*60}")
        print(f"📚 SOURCES ({result['num_docs_used']} docs used from {result['num_docs_retrieved']} retrieved):")
        sources_seen = set()
        for doc in result["source_documents"]:
            src = doc.metadata.get("source_file", "Unknown")
            page = doc.metadata.get("page", "N/A")
            year = doc.metadata.get("year", "?")
            cat = doc.metadata.get("category", "?")
            key = f"{src}-{page}"
            if key not in sources_seen:
                sources_seen.add(key)
                print(f"  📄 {src} | Page {page} | Year: {year} | Category: {cat}")
        print('─'*60)

    # ──────────────────────────────────────────────────────
    # Evaluation Method
    # ──────────────────────────────────────────────────────
    def evaluate(
        self,
        test_questions: List[str],
        ground_truths: Optional[List[str]] = None,
    ):
        """
        Runs Ragas evaluation on a list of test questions.

        Example:
            pipeline.evaluate(
                test_questions=["What is BERT?", "How does TF-IDF work?"],
                ground_truths=["BERT is...", "TF-IDF stands for..."]
            )
        """
        print("\n📊 RUNNING RAGAS EVALUATION")
        print("="*60)

        questions, answers, contexts = [], [], []

        for q in test_questions:
            result = self.query(q, verbose=False)
            questions.append(q)
            answers.append(result["answer"])
            contexts.append([doc.page_content for doc in result["source_documents"]])

        scores = evaluate_with_ragas(questions, answers, contexts, ground_truths)

        if scores:
            print("\n📈 RAGAS SCORES:")
            print(f"  Faithfulness:      {scores.get('faithfulness', 'N/A'):.3f}  (1.0 = no hallucination)")
            print(f"  Answer Relevancy:  {scores.get('answer_relevancy', 'N/A'):.3f}  (1.0 = perfectly on-topic)")
            print(f"  Context Precision: {scores.get('context_precision', 'N/A'):.3f}  (1.0 = all context is useful)")
            if ground_truths and ground_truths[0]:
                print(f"  Context Recall:    {scores.get('context_recall', 'N/A'):.3f}  (1.0 = nothing missed)")

        return scores
```
